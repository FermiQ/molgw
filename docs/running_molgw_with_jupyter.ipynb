```python
from molgw import Molgw_outputs, Molgw_input
import numpy as np
import matplotlib.pyplot as plt
```

```python
parameters = { 'scf': 'hf',
              'basis': 'cc-pVTZ',
              'auxil_basis': 'auto'
             }
xxx = np.arange(0.3,2.01,0.05)
```

```python
yyy = []
for x in xxx:
    structure = f"H {x} 0. 0.\n" + "H 0. 0. 0.\n"
    parameters["xyz"] = structure
    mlgi = Molgw_input(parameters)
    mlgo = mlgi.run(tmp="tmp/")
    yyy.append(mlgo["scf energy"]["total"])
```

```python
plt.plot(xxx,yyy,color='red',label='Hartree-Fock')
plt.legend()
plt.xlabel('H-H (Angstrom)')
plt.ylabel('Energy (Ha)')
plt.show()
```
